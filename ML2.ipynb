```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/dataset/citrus.csv')
```

```python
df.head(10000)
```

**MELIHAT INFORMASI DATAFRAME**

```python
df.info()
```

**MENGUBAH TEXT MENJADI NUMERIK**

```python
df.name[df.name == 'orange'] = 0
df.name[df.name == 'grapefruit'] = 1
```

**MENGUBAH VALUE**

```python
dataset = df.values
```

```python
dataset
```

**MENGAMBIL KOLOM 1 SAMPAI 6 KARENA ITU YANG BERISI ATRIBUT**

```python
X = dataset[:,1:6]
```

```python
y = dataset[:,0]
```

**PROSES NORMALISASI DATAFRAME**

```python
from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler()
X_scale = min_max_scaler.fit_transform(X)
X_scale
```

**MELAKUKAN PELATIHAN DAN PENGUJIAN**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test, = train_test_split(X_scale, y, test_size=0.3)
```

**MENGUBAH LABEL MENJADI TIPE DATA BOLEAN FLOAT 32, KARENA MENGUBAH NILAI PADA KOLOM NAME**

```python
import numpy as np

Y_train = Y_train.astype(np.float32)
Y_test = Y_test.astype(np.float32)
```

**MEMBANGUN MODEL**

```python
from keras.models import Sequential
from keras.layers import Dense
```

**MEMBUAT 3 LAYER**

```python
model = Sequential([
                    Dense(32, activation='relu', input_shape=(5,)),
                    Dense(32, activation='relu'),
                    Dense(1, activation='sigmoid'),])
```

**MENAMBAHKAN PARAMETER AKURASI**

```python
model.compile(optimizer='sgd',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

**MELAKUKAN PELATIHAN MODEL**

```python
model.fit(X_train, Y_train, epochs=100)
```

**MELAKUKAN EVALUASI**

```python
model.evaluate(X_test, Y_test)
```
