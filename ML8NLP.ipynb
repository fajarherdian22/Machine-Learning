**import library tokenizer**

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

**pengubahan objek kata menjadi token**

```python
tokenizer = Tokenizer(num_words=15, oov_token='-')
```

**masukan kata yang akan di tokenizing**

```python
teks = ['saya suka programming',
        'Programming sangat menyenangkan!',
        'Machine Learning berbeda dengan pemrograman konvensional']
```

**proses tokenisasi**

```python
tokenizer.fit_on_texts(teks)
```

**mengubah teks yang dibuat ke sequences**

```python
sequences = tokenizer.texts_to_sequences(teks)
```

**menampilkan kata berdasarkan token**

```python
print(tokenizer.word_index)
```

**menambahkan pad sequences(untuk membuat memiliki panjang yang seragam)**

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
sequences_samapanjang = pad_sequences(sequences)
```

**menampilkan pad sequence**

```python
print(sequences_samapanjang)
```

**memindahkan angka 0 menjadi diakhir**

```python
sequences_samapanjang = pad_sequences(sequences, 
                                      padding='post',
                                      maxlen=5,
                                      truncating='post')
```

```python
print(sequences_samapanjang)
```
